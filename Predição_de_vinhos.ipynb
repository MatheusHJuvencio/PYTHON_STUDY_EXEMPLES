#### Exercício de predição de vinhos tinto e/ou branco

```python
# carregando os dados
import pandas as pd
df = pd.read_csv('data/wine_dataset.csv')
df.head()
```

```python
# gostariamos de classificar os vinhos em tinto (red) ou branco (white)
# porém, o pandas aceita apenas numeros, logo, trocar red por 0 (zero)
# e white por 1 (um).
# trocando apenas na coluna style e usando a função raplace
df['style'] = df['style'].replace('red', 0)
df['style'] = df['style'].replace('white', 1)
```

```python
# conferindo se alterou a coluna style
df.head()
```

```python
# gostariamos de prever se é tinto ou branco, assim, nossa variável alvo (Y) é a coluna style
# o preditores é a variável (x) é todo o resto dos dados, logo, removendo a coluna style (variável alvo)
y = df['style']
x = df.drop('style', axis = 1) #axis=1 (colunas)
print('A variável y é: \n%s\n'%y)
print('A variável x é: \n%s\n'%x)
```

```python
# no machine learning, importante separar os dados em conjuntos de treinos e teste
# assim, utilizamos o sklearn para tal função
from sklearn.model_selection import train_test_split

#criando os conjuntos de datos de treino e teste:
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.3) # tamanho especifico para teste, no caso, 30%
```

```python
# agora, após os tratamentos dos dados, utilizaremos o algoritmo de ML ExtraTreeClassifier
from sklearn.ensemble import ExtraTreesClassifier

# criação do modelo:
modelo = ExtraTreesClassifier() # chamada da função
modelo.fit(x_treino, y_treino) # treinar o algoritmo

# imprimindo os resultados:
resultado = modelo.score(x_teste, y_teste) # comparou os teste com os resultados do treino 
print('Acurácia: \n%s\n'%resultado)
```

```python
# fazendo predição
previsao = modelo.predict(x_teste[500:510])
print('a previsão é: \n%s\n'%previsao)
```
