````python
# leitura dos dados:

data = open('data/u.data')
dados = data.read().splitlines()
````

````python
# cabeçalho

headers = ['user_id', 'item_id', 'rating', 'timestamp']
````

````python
# função para formatar linhas:

def formatar_linhas(headers: list, lista:list, lista_dicionario:list):
    
    dicionario_temporario = {}
    for linha in lista:
        for h in range(len(headers)):
            l = linha.split('\t')
            #criação de dicionário
            dicionario_temporario[headers[0]] = l[0]
            dicionario_temporario[headers[1]] = l[1]
            dicionario_temporario[headers[2]] = int(l[2])
            dicionario_temporario[headers[3]] = l[3]
        lista_dicionario.append(dicionario_temporario.copy())
        dicionario_temporario.clear()
````

````python
# chamando a função
lista_dicionario = []

formatar_linhas(headers, dados, lista_dicionario)
````

````python
# media de todas as avaliações:

soma = 0

for item in lista_dicionario:
    soma = soma + item['rating']
media_rating = soma/len(lista_dicionario)

print(f'A média de todos os filmes é: {media_rating}')
````

````python
# Leitura dos dados:
data_user = open('data/u.user')
user = data_user.read().splitlines()

usuarios_unicos = []

for linha in user:
    l = linha.split("|")
    usuarios_unicos.append(l[0])
    
print(len(usuarios_unicos))
````

````python
# Função para cálculo da media por id:

def id_media(lista_entrada: list, lista_ids: list, lista_saida: list):

    soma = 0
    cont = 0
    dict_temp = {}

    for user in lista_ids:
        for item in lista_entrada:
            if item['user_id'] == user:
                soma = soma + item['rating']
                cont = cont +1
        dict_temp['user_id'] = user
        dict_temp['media'] = (soma/cont)
        dict_temp['qtd_votos'] = cont
        lista_saida.append(dict_temp.copy())
        dict_temp.clear()
        soma = 0
        cont = 0
        
````

````python
media_user_rating = []

id_media(lista_dicionario, usuarios_unicos, media_user_rating)
print(media_user_rating[:10])
````

## Exercício 2: Filmes mais mal avaliados

````python
# leitura do arquivo u.item

u_item = open('data/u.item', encoding='iso-8859-1') 
filme_data = u_item.read().splitlines()
print(filme_data[0])
````

````python
# lista com os ids dos filmes
ids_filmes = [] 
# lista com os titulos dos filmes
filmes = []  
# lista com os anos de cada filme
years = []   
# lista com os generos dos filmes
genero_filmes = []

dicionario_temporario = {}

for linha in filme_data:
    
    l = linha.split("|")
    date = l[2].split("-")
    
    ids_filmes.append(l[0])
    
    filmes.append(l[1])
    
    years.append(date[len(date)-1])
    
    dicionario_temporario['Unknown'] = l[5]
    dicionario_temporario['Action'] = l[6]
    dicionario_temporario['Adventure'] = l[7]
    dicionario_temporario['Animation'] = l[8]
    dicionario_temporario['Children'] = l[9]
    dicionario_temporario['Comedy'] = l[10]
    dicionario_temporario['Crime'] = l[11]
    dicionario_temporario['Documentary'] = l[12]
    dicionario_temporario['Drama'] = l[13]
    dicionario_temporario['Fantasy'] = l[14]
    dicionario_temporario['Film_Noir'] = l[15]
    dicionario_temporario['Horror'] = l[16]
    dicionario_temporario['Musical'] = l[17]
    dicionario_temporario['Mystery'] = l[18]
    dicionario_temporario['Romance'] = l[19]
    dicionario_temporario['Sci_Fi'] = l[20]
    dicionario_temporario['Thriller'] = l[21]
    dicionario_temporario['War'] = l[22]
    dicionario_temporario['Western'] = l[23]
    
    genero_filmes.append(dicionario_temporario.copy())
    dicionario_temporario.clear()
    
print(f'Quantidade de filmes: {len(ids_filmes)}')
print(f'Primeiros 5 titulos da lista: {filmes[:5]}')
print(f'Primeiros 5 generos da lista: {genero_filmes[:5]}')
````

````python
def itemid_media(lista_entrada: list, ids_filmes: list, lista_saida: list):
    
    soma = 0
    cont = 0
    dicionario_temporario = {}
    
    for user in ids_filmes: 
        for item in lista_entrada:
            if item['item_id'] == user:
                soma += item['rating']
                cont += 1
        dicionario_temporario['item_id'] = user
        dicionario_temporario['média'] = (soma / cont)
        lista_saida.append(dicionario_temporario.copy())
        dicionario_temporario.clear()
        soma = 0
        cont = 0
````

````python
# chamando a função

media_item_rating = []
itemid_media(lista_dicionario, ids_filmes, media_item_rating)

print(media_item_rating[:5])
````

````python
media_item = []

for dupla in media_item_rating:
    m = dupla['média']
    media_item.append(m)
````

````python
media_item.sort()

print(media_item[:5])
print('--'*45)
print('Filmes mais mal avaliados: ', end=' ')

for dupla in media_item_rating:
    if media_item[0] == dupla['média']:
        num = int(dupla['item_id'])
        print(filmes[num-1], end = '; ')
````

## Exercício 4 - Média de avaliação para cada gênero de filmes

````python
print(filme_data[len(filme_data)-1])    

header_genero = ['unknown', 'Action', 'Adventure', 
                 'Animation', 'Children', 'Comedy', 
                 'Crime', 'Documentary', 'Drama', 
                 'Fantasy', 'Film_Noir', 'Horror', 
                 'Musical', 'Mystery', 'Romance', 
                 'Sci_Fi', 'Thriller', 'War', 'Western']
````

````python
itemid = []
genero_itemid = {}

for i, genero in enumerate(header_genero):
    for linha in filme_data:
        l = linha.split('|')
        if l[i+5] == '1':
            itemid.append(l[0])
    genero_itemid[genero] = itemid[:]
    itemid.clear()
    
print(genero_itemid)
````

````python
soma = cont = 0

genero_medias = {}

for key, value in genero_itemid.items():
    for dupla in media_item_rating:
        for v in value:
            if v == dupla['item_id']:
                soma += dupla['média']
                cont += 1
    media = soma/cont
    genero_medias[key] = media
    soma = cont = 0
````

````python
print('GENERO          RATING MÉDIO')
print('-'*30)
for k, v in genero_medias.items():
    print(f'{k:<20} {v:.2f}')
````

## Exercício 5 - Avaliação média de filmes por ano

````python
itemid_year = []


for linha in filme_data:
    l = linha.split("|")
    complete_date = l[2].split("-")
    year = complete_date[len(complete_date)-1]
    dicionario_temporario['item_id'] = l[0]
    dicionario_temporario['ano'] = year
    itemid_year.append(dicionario_temporario.copy())
    dicionario_temporario.clear()

print(itemid_year[:5])
print(years[:5])
````

````python
item = []
year_items = {}

for y in years:
    for dupla in itemid_year:
        if dupla['ano'] == y:
            item.append(dupla['item_id'])
    year_items[y] = item[:]
    item.clear()

print(year_items)
````

````python
soma = cont = 0
year_medias = {}

for key, value in year_items.items():
    for par in media_item_rating:
        for v in value:
            if v == par['item_id']:
                soma += par['média']
                cont += 1
    media = soma / cont
    year_medias[key] = media
    soma = cont = 0

print(year_medias)
````

````python
from operator import itemgetter

resultado5 = sorted(year_medias.items(), key=itemgetter(1))

# itemgetter: 0 ordenada a chave; 1 ordena o valor 
# PS: ao usar esta função, a saída vira uma lista

print('\033[33mANO COM MAIOR MÉDIA DE RATING:\033[m')
print(f'ANO: {resultado5[len(resultado5)-1][0]}    RATING MÉDIO {resultado5[len(resultado5)-1][1]:.2f}')
````

````python
with open('data/u.user', 'r') as file: user_data = file.read().splitlines()
with open('data/u.occupation', 'r') as file_occupation: occupations = file_occupation.read().splitlines()

print(user_data[:5])
print(occupations)
````

````python
userid_occupation = []
temporario = []
for linha in user_data:
    l = linha.split('|')
    usr = l[0]
    ocp = l[3]
    temporario.append(usr)
    temporario.append(ocp)
    userid_occupation.append(temporario[:])
    temporario.clear()

print(userid_occupation[:5])
````

````python

occupation_users = {}

for occup in occupations: 
    for dupla in userid_occupation:
        if occup == dupla[1]:
            temporario.append(dupla[0])
    occupation_users[occup] = temporario[:]
    temporario.clear()


print(occupation_users)
````

````python
soma = cont = 0
occupation_medias = {}

for key, value in occupation_users.items(): 
    for dupla in media_user_rating:
        for v in value:
            if v == dupla['user_id']:
                soma += dupla['media']
                cont += 1
    media = soma / cont
    occupation_medias[key] = media
    soma = cont = 0 

print(occupation_medias)
````
